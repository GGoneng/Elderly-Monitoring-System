```python
# -----------------------------------------------------------------------------------
# 파일명       : Data_Preprocessing.ipynb
# 설명         : 시계열 데이터를 Sliding Window 방식으로 패턴화     
# 작성자       : 이민하
# 작성일       : 2024-11-12
# 
# 사용 모듈    :
# - pandas       # 데이터프레임 기반 데이터 처리
# -----------------------------------------------------------------------------------
# >> 주요 기능
# - Sliding Window 알고리즘으로 데이터 패턴화
# - 모델별로 최적의 데이터 형태로 변형
#
# >> 업데이트 내역
# [2024-11-12] 일주일 단위의 Window로 패턴화
# [2024-11-13] 2주일 단위의 Window로 패턴화
# [2024-11-28] 일주일 + 1/4일 단위의 Window로 패턴화 (이상 데이터 제거)
# [2024-12-05] 일주일 단위의 Window로 데이터 변화량 패턴화 (이상 데이터 제거)
# [2024-12-08] 일주일 단위의 Window로 패턴화 (이상 데이터 제거)
# -----------------------------------------------------------------------------------
```

```python
# 데이터프레임 기반 데이터 처리
import pandas as pd
```

```python
# 데이터 경로 설정
DATA_PATH = '../Data/수도전기사용량_3개월분_수정본.xlsx'

# 이상 데이터가 제거된 엑셀 파일 -> 데이터프레임으로 불러오기
df1 = pd.read_excel(DATA_PATH, header = 1)
df2 = pd.read_excel(DATA_PATH, header = 1, sheet_name = '수도사용량_3개월분')

df1
```

```python
df2
```

```python
# 결측치 처리
df1.drop(1001, axis = 0, inplace = True)

df1.dropna(axis = 1, inplace = True)
df2.dropna(axis = 1, inplace = True)

df1
```

```python
# 불필요한 컬럼 제거 (0 컬럼)
df1.drop(df1.columns[0], axis = 1, inplace = True)
df2.drop(df2.columns[0], axis = 1, inplace = True)
```

```python
# 6시간 단위의 이름 리스트 생성
column_name = pd.date_range('2023-06-01', '2023-09-01', freq = '6H')
column_name = column_name[:-1]

column_name
```

```python
# 6시간 단위의 이름으로 컬럼명 변경
df1.columns = column_name
df2.columns = column_name

df1.columns
```

```python
df1[df1.columns[0:12]]
```

```python
# 불필요한 행 삭제 (0 행)
df1.drop(0, axis = 0, inplace = True)
df2.drop(0, axis = 0, inplace = True)

df1
```

```python
# 1000가구의 3달치 데이터 -> 가구 구분 없이 일주일 단위의 패턴으로 변경
def slicing(data, step, max_length, num):

    # 빈 데이터프레임 생성 (열 갯수는 step 만큼)
    df = pd.DataFrame(columns = range(0, step))

    # 다음 행으로 이동하기 위한 index
    diff = max_length - step

    # List Slicing을 활용해 데이터 변경
    for i in range(num):
        for j in range(diff):
            df.loc[i * diff + j] = list(data.iloc[i][data.columns[j : j + step + 1]])
    
    return df
```

```python
# 전기 데이터 -> 일주일 단위의 패턴으로 변경
electric_df = slicing(df1, 28, 368, 1000)
electric_df
```

```python
# CSV 파일로 저장
electric_df.to_csv('../Data/electric_df_28_days.csv')
```

```python
# 수도 데이터 -> 일주일 단위의 패턴으로 변경
water_df = slicing(df2, 27, 368, 1000)
water_df
```

```python
# CSV 파일로 저장
water_df.to_csv('../Data/water_df_28_days.csv')
```
